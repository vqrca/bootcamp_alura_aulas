```python
dados = pd.read_excel('https://github.com/vqrca/bootcamp_alura_aulas/blob/main/Modulo_4/Kaggle_Sirio_Libanes_ICU_Prediction.xlsx?raw=true')
dados.head()
```

Vamos definir as features com valores contínuos, que são as medições colunas 13 a -2:

```python
dados.iloc[:, 13:-2].columns
```

Agora vamos criar uma função para o preenchimento das linhas das features contínuas sem valores, usando os métodos `bfill` e `ffill`.

Além disso, vamos definir as features categóricas e os dados de saída:

- categóricas:  `dados.iloc[:, :13]` (São as colunas com valores de 0 e 1)
- saída: `dados.iloc[:, -2:]`

Posteriormente, os dados serão concatenados `pd.concat([features_categoricas, features_continuas, saida]`

```python
def preenche_tabela(dados):
    features_continuas_colunas = dados.iloc[:, 13:-2].columns
    features_continuas = dados.groupby("PATIENT_VISIT_IDENTIFIER", as_index=False)[features_continuas_colunas].fillna(method='bfill').fillna(method='ffill')
    features_categoricas = dados.iloc[:, :13]
    saida = dados.iloc[:, -2:] #:, todas as linhas
    dados_finais = pd.concat([features_categoricas, features_continuas, saida], ignore_index=True,axis=1) #definir eixo=1 
    dados_finais.columns = dados.columns
    return dados_finais
```

```python
dados_limpos = preenche_tabela(dados)
dados_limpos.head()
```

```python
dados_limpos.describe()
```

Agora vemos que todos os valores de count estão próximos, exceto nas colunas de doenças. Há algum paciente que não tem as doenças e algumas outras coisas. Então, vamos precisar dar um `dropna`:

```python
x_columns = dados.describe().columns

dados_limpos = dados_limpos.dropna()
```

Agora vamos retirar pacientes que entraram na UTI nas duas primeiras horas:

```python
a_remover = dados_limpos.query("WINDOW=='0-2' and ICU==1")['PATIENT_VISIT_IDENTIFIER'].values
dados_limpos = dados_limpos.query("PATIENT_VISIT_IDENTIFIER not in @a_remover")
dados_limpos.describe()
```

Agora precisamos definir qua janela pegar!

Vamos simplificar e pegar a primeira janela: quando a pessoa entra (nas duas primeiras horas). 

Porém, precisamos saber se o paciente foi ou não para UTI:

Para isso foi criada uma função que marca a primeira linha e devolve a primeira linha, sendo todos os pacientes (uma linha por paciente) com a informação se ele foi para UTI (=1) e se ele não foi para UTI (=0).

```python
def prepare_window(rows):
    if(np.any(rows["ICU"])):
        rows.loc[rows["WINDOW"]=="0-2", "ICU"] = 1
    return rows.loc[rows["WINDOW"] == "0-2"]

"""Essa função recebe as linhas (rows), que são as linhas de um paciente e se qualquer
linha ICU do paciente for = 1  vamos pegar a primeira linha desse paciente e marcar que
ele foi para UTI. Então, a linha onde está a janela 0-2 será localizada e retornada através de 
rows.loc[rows["WINDOW"] == "0-2" """

dados_limpos = dados_limpos.groupby("PATIENT_VISIT_IDENTIFIER").apply(prepare_window)
dados_limpos.head()
```

Separar os dados em x e y, para aplicar o modelo:

```python
y = dados_limpos["ICU"]
x = dados_limpos[x_columns].drop(["ICU"], axis=1)
```

Adicionando o seed, para que o `testsplit` ocorra da mesma forma:

```python
np.random.seed(79230)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y)
```

DummyClassifier

```python
modelo = DummyClassifier()
modelo.fit(x_train, y_train)

y_prediction = modelo.predict(x_test)
accuracy_score(y_test, y_prediction)
```

LogisticRegression

```python
modelo = LogisticRegression()
modelo.fit(x_train, y_train)

y_prediction = modelo.predict(x_test)
accuracy_score(y_test, y_prediction)
```

###**Desafio 03 -pesquisar aprendizado supervisionado, não supervisionado, por reforço e mais**

**Aprendizado supervisionado:**  Nesse caso, estamos ensinando à máquina como trabalhar com os nossos dados, logo, estamos supervisionando o aprendizado da máquina. 

No aprendizado supervisionado passamos todos os dados para treinar o modelo. Isto é, passamos os dados e os resultados. Com base nisso, o algoritmo descobre uma forma de trabalhar com os dados.

**Aprendizado não supervisionado:** Diferente do aprendizado supervisionado, neste tipo de aprendizado nós não passamos os resultados para ensinar o algoritmo. 

O próprio algoritmo pega o conjunto de dados e descobre como trabalhar com eles. Os tipos de algoritmos mais utilizados nesse tipo de aprendizado são o clustering e as regras de associação. Um exemplo de algoritmo é o K-Means.

**Aprendizado de reforço:** Nele a máquina interage com o ambiente. Essa interação gera um feedback, positivo ou negativo. Caso seja um feedback positivo, a máquina grava que pode realizar essa ação novamente, em caso de feedback negativo, o modelo sabe que não pode executá-lo novamente.

###**Desafio 06 - Pesquisar sobre ética em data science/ciência de dados e inteligência artificial.**

A ciência de dados oferece grandes oportunidades para melhorar a vida privada e pública, bem como o nosso meio ambiente.Porém, essas oportunidades também estão associadas a desafios éticos significativos. O uso extensivo de cada vez mais dados - muitas vezes pessoais, senão confidenciais e a crescente dependência de algoritmos para analisá-los a fim de moldar escolhas e tomar decisões, também como a redução gradual do envolvimento humano ou mesmo a supervisão de muitos processos automáticos, colocam questões prementes de justiça, responsabilidade e respeito aos direitos humanos, entre outros.

A [Lei Geral de Proteção de Dados Pessoais (LGPD)](http://www.planalto.gov.br/ccivil_03/_ato2015-2018/2018/lei/L13709.html) entrou em vigor em setembro de 2020 no Brasil e regulamenta qualquer atividade que envolva utilização de dados pessoais. É aplicável a toda operação de tratamento de dados realizada por pessoa física ou jurídica, pública ou privada, independente do meio e de onde os dados estejam localizados. Na prática, a LGPD busca dar às pessoas mais controle sobre quais dados são fornecidos às empresas e como eles são utilizados.

Com a LGPD, regulamenta-se o controle dos dados pelos usuários, tornando obrigatório que as empresas informem quais informações são coletadas e como elas serão usadas. Assim, as pessoas têm a opção de autorizar ou não o compartilhamento.

Algoritmos de inteligência artificial são beneficiados pela maior quantidade de dados possível para a detecção de padrões, e é desejável mantê-los armazenados, já que os padrões históricos podem servir como base para melhores decisões das IAs. É aí que está o grande desafio da LGPD: é preciso limitar a coleta de dados àqueles necessários para os objetivos declarados, reduzindo também o tempo de armazenamento deles, e declarando todas essas questões de forma transparente aos usuários.

Uma vez que os dados foram coletados, a empresa precisa ser capaz de alterar ou apagar os dados do usuário caso ele faça essa solicitação. Portanto, para estar em conformidade com a LGPD, os dados usados em sistemas de inteligência artificial precisarão ser identificáveis e acessíveis a nível individual.

Além disso, é possível que a proteção de dados incentive o desenvolvimento de inteligências artificiais que não dependem necessariamente dos dados compartilhados pelos próprios usuários, como é o caso dos sistemas de reconhecimento de imagem, nos quais a LGPD terá menor influência.

Tecnologias de inteligência artificial não serão prejudicadas, se administradas seguindo o que determina a lei: informar os dados coletados e o tempo de armazenamento, certificar-se de que sejam identificáveis e removíveis, e explicar o que será feito com as informações.
